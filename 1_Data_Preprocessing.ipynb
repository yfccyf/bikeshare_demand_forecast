```python
# rename columns
df.rename(columns={"Rented Bike Count": "count", "Hour": "hour", "Temperature(°C)": "temp", "Humidity(%)": "humidity", "Wind speed (m/s)": "wind_speed", "Visibility (10m)": "visibility", "Dew point temperature(°C)": "dew_point_temp", "Solar Radiation (MJ/m2)": "solar_radiation", "Rainfall(mm)": "rainfall", "Snowfall (cm)": "snowfall", "Seasons": "seasons", "Holiday": "holiday", "Functioning Day": "functioning_day"}, inplace=True)
```

##### Combine date and hour columns to datetime

```python
df = df.reset_index()
```

```python
df.head()
```

```python
# assemble a new "datetime" column
df['datetime'] = pd.to_datetime(dict(year=df['Date'].dt.year, month=df['Date'].dt.month, day=df['Date'].dt.day, hour=df['hour']))
```

```python
df.head()
```

```python
# drop the previous Date column
df.drop(columns=['Date'], inplace=True)
```

```python
df.drop(columns=['hour'], inplace=True)
```

```python
df.head()
```

##### Extract date parts

```python
df['date'] = df['datetime'].dt.date
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
df['hour'] = df['datetime'].dt.hour
```
